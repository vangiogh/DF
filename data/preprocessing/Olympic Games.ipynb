```python
import pandas as pd

df = pd.read_csv('../raw/olympic_all.csv', index_col=0)
df.Medal = pd.Categorical(df.Medal, categories=['Bronze', 'Silver', 'Gold'], ordered=True)
df = df.query('Season == "Summer" & ~Medal.isnull()')
df = df.sort_values(['Year', 'City', 'Sport', 'Event'])
df
```

```python
columns = ['City', 'Year', 'Sport', 'Event', 'Sex', 'Medal', 'Age']
```

```python
df1 = df[columns + ['Name', 'NOC']]
```

```python
df1.to_csv('../raw/olympic_champions.csv')
```

```python
df = df[columns].value_counts(sort=False).sort_index(level=['Year', 'Sport', 'Medal']).reset_index(name='Count')
```
